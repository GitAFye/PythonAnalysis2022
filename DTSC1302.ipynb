```python
import csv #import csv package

fileobject = open("2011-2021 DATA.csv",'r') #open file to read
all_data = csv.reader(fileobject) #read file and store in "all_data" instance
data_list=list(all_data) #convert lines in all_data instance as entries of a list
print(data_list)



for i in range(len(data_list)):
    for j in range(len(data_list[i])):
        if i > 0 and j > 0:
            try:
                data_list[i][j] = float(data_list[i][j])
            except:
                print(str(data_list[i][j]))
print(data_list)
```

```python
#create and store lists for all data per variable 
year=data_list[0][1:] #store list of header strings as "year"
print(year)
cpi=data_list[1][1:]
print(cpi)
cost_liv=data_list[2][1:]
print(cost_liv)
afford=data_list[3][1:]
print(afford)
med_income=data_list[4][1:]
print(med_income)
assist=data_list[5][1:]
print(assist)
retirement=data_list[6][1:]
print(retirement)
upper_pop=data_list[7][1:]
print(upper_pop)
poverty_pop=data_list[8][1:]
print(poverty_pop)
```

```python
#import libraries needed
import matplotlib.pyplot as plt
import statistics
```

```python
plt.plot(year,med_income)
plt.xlabel("Year")
plt.ylabel("Median Income in $")
plt.title("Charlotte Median Income over 10 Years")
plt.xticks(year,rotation=90)
plt.ylim(60000,75000)
plt.show()
print("The range of our Median Income data goes from the minimum value of " + str(int(min(med_income))) + " " + "to the maximum value of " + str(int(max(med_income)))+ ".")
medinc_minpos=med_income.index(min(med_income))
#print(medinc_minpos)
medinc_maxpos=med_income.index(max(med_income))
#print(medinc_maxpos)
print("Of the years we explored, Charlotte's reported median income is lowest in the year " + str(year[0]) + " and highest in the year " + str(year[9]) + ".")
medinc_change= max(med_income)-min(med_income)
medinc_changeperc= 100*(medinc_change/float(med_income[0]))
print("The total change in median income over ten years is " + str(medinc_change) + " dollars. This is about a " + str(int(medinc_changeperc)) + "% increase from it's starting value.")
```

```python
plt.plot(year, assist)
plt.xlabel("Year")
plt.ylabel("Public Cash Assistance in $")
plt.title("Public Cash Assistance Over the Years")
plt.xticks(year, rotation=90)
plt.show()
assist_change= assist[0]-assist[-1]
assist_changeperc= 100*(assist_change/assist[0])
print("The change in Cash Assistance income received by Charlotte residents is "+ str(assist_change)+ " dollars.This is about a " + str(int(assist_changeperc)) + "% change from it's starting value.")
```

```python
plt.plot(year, afford)
plt.xlabel("Year")
plt.ylabel("Surplus income in $")
plt.xticks(year , rotation = 90)
plt.title("Affordability of Charlotte")
plt.show()
print("The range of our Affordability measurement goes from the minimum value of " + str(int(min(afford))) + " " + "to the maximum value of " + str(int(max(afford)))+ ".")
afford_minpos=afford.index(min(afford))
#print(afford_minpos)
afford_maxpos=afford.index(max(afford))
#print(afford_maxpos)
print("Charlotte was most affordable in the year " + str(year[5]) + " and least affordable in the year " + str(year[9]) + ".")
afford_roc= (afford[0]-afford[-1])/10
print("The rate of change for the affordability of Charlotte is a decrease of approximately " + str(int(afford_roc)) + " dollars per year.")
```

```python
plt.plot(year, retirement)
plt.xlabel("Year")
plt.ylabel("Income in $")
plt.xticks(year , rotation = 90)
plt.title("Total Retirement Income")
plt.show()
print("The range of total retirement income goes from the minimum value of " + str(int(min(retirement))) + " " + "to the maximum value of " + str(int(max(retirement)))+ ".")
retirement_minpos=retirement.index(min(retirement))
#print(retirement_minpos)
retirement_maxpos=retirement.index(max(retirement))
#print(retirement_maxpos)
print("Charlotte's retired population experienced its lowest income in the year " + str(year[2]) + " and its highest income in the year " + str(year[6]) + ".")
retire_change= retirement[-1]-retirement[0]
retire_changeperc= 100*(retire_change/retirement[0])
print("The change in Total Retirement income for Charlotte residents is "+ str(retire_change)+ " dollars.This is about a " + str(int(retire_changeperc)) + "% change from it's starting value.")
```

```python
plt.plot(year, cost_liv)
plt.xlabel("Year")
plt.ylabel("Cost of Living in $")
plt.title("Charlotte Cost of Living Over Ten Years")
plt.xticks(year, rotation=90)
plt.show()
print("The range of our Cost of Living data goes from the minimum value of " + str(int(min(cost_liv))) + " " + "to the maximum value of " + str(int(max(cost_liv))) + ".")
cost_minpos=cost_liv.index(min(cost_liv))
#print(cost_minpos)
cost_maxpos=cost_liv.index(max(cost_liv))
#print(cost_maxpos)
print("Cost of living is lowest in the year " + str(year[0]) + " and highest in the year " + str(year[9]) +".")
costliv_change= max(cost_liv)-min(cost_liv)
costliv_changeperc= 100*(costliv_change/float(cost_liv[0]))
print("The total change in cost of living over ten years is " + str(costliv_change) + " dollars. This is about a " + str(int(costliv_changeperc)) + "% increase from it's starting value.")
```

```python
plt.plot(year, upper_pop)
plt.xlabel("Year")
plt.ylabel("Population Percent")
plt.title("Charlotte's Population with Income of at least $200k")
plt.xticks(year, rotation=90)
plt.show()
upper_change= max(upper_pop)-min(upper_pop)
print("The range of the percentage of Charlotte's population goes from the minimum value of " + str(float(min(upper_pop))) + " % to the maximum value of " + str(float(max(upper_pop))) + " %. This subpopulation experienced an increase of " + str(upper_change) + "%.")
```

```python
plt.plot(year, poverty_pop)
plt.xlabel("Year")
plt.ylabel("Population Percent")
plt.title("Charlotte's Population Below Poverty Line")
plt.xticks(year, rotation=90)
plt.show()
poverty_change= max(poverty_pop)-min(poverty_pop)
print("The range of the percentage of Charlotte's population below the poverty line goes from the minimum value of " + str(float(min(poverty_pop))) + " % to the maximum value of " + str(float(max(poverty_pop))) + " %. This subpopulation experienced a decrease of " + str(poverty_change) + "%.")
```
